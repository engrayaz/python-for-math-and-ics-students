```python
What is a Function?
A function is a block of code that does a specific job.
Instead of writing the same code again and again, we use a function to do that work for us.
“Just like a calculator has buttons for add, subtract, multiply. 
You press a button and the work is done. In Python, functions work in the same way.”
What are Built-in Functions?
“Built-in functions are the functions that are already provided by Python.
We do not need to create them. We just use them.”
print() → shows output
range() → gives numbers for loops

We have already been using functions like print() and range() 
without knowing they are called functions.
```

```python
How to Use a Function?
General idea:
function_name(value)
Explain:
We write the function name
We give data inside brackets
Python does the work for us
```
