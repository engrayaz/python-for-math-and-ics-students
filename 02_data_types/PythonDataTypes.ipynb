 What are Data Types?
As you know, when we create a variable in Python, the value is stored somewhere in the computer’s memory.For example:
x = 5
Here:
x is the variable name
5 is the value stored in memory
In Python, a data type tells us what kind of value a variable holds, such as int, float, str, bool, and complex.
Common Built-in Data Types:
int → whole numbers (10, -3)
float → decimal numbers (2.5, 0.001)
str → text ("Python")
bool → True or False
complex → complex numbers (2 + 3j)

```python
x=5
print(x)
```

```python
y=-3.5
print(y)
```

```python
Name_lang="Python"
print(Name_lang)
```

```python
z=5+12j
print(z)
```

Summary
Variables store values in memory
Data types tell us what kind of value it is
Common types: int, float, str, bool, complex
