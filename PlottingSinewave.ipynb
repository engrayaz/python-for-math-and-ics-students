````python
import math
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(0,359):
  angle=math.radians(i)
  sinV=math.sin(angle)
  x.append(angle)
  y.append(sinV)
plt.plot(x,y)
plt.show()
````

````python
import math
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(0,359):
  angle=math.radians(i)
  sinV=3*math.sin(angle)
  x.append(angle)
  y.append(sinV)
plt.plot(x,y)
plt.show()
````
