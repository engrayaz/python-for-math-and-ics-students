```python
Simple definition: A while loop keeps running while a condition is true. 
When the condition becomes false, the loop stops.

condition is checked again and again
If condition is True → loop runs
If condition is False → loop stops
The indented lines are the body of the loop
```

```python
General Syntax
while condition:
    statements
```

```python
#Example1: Counting 1 to 5
i=1
while i<=5:
    print(i)
    i=i+1
```

```python
#example2
count=1
while count<=3:
    print("Ali")
    count=count+1
```

```python
i=3
while i<=30:
    print(i)
    i=i+3
```

```python
Summary Today, we learned about the while loop in Python. 
The while loop runs as long as a condition is true. 
We learned how to use a counter variable and how to stop the loop by changing the variable. 
We also learned that forgetting to change the variable can cause an infinite loop. 
The body of the while loop must be properly indented.

Comparing for and while

“We use for loop when we know exactly how many times to repeat.
We use while loop when we want to repeat until a condition becomes false.”
Example idea (no need to code both):
for → run 5 times
while → run while something is true
```
