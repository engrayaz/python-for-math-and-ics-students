Types of Data Structures in Python
List, tuple,dictionary,set etc

```python
list_a=[10,1.5,"Ali",True,3-5j]
print(list_a)
```

```python
print(type(list_a))
```

```python
print(len(list_a))
```

```python
tuple1=("john",2,True,3+5j)
print(tuple1)
print(type(tuple1))
print(len(tuple1))
```

```python
student = {"name": "Ali", "marks": 85, "passed": True}
print(student)
print(type(student))
print(len(student))
```

```python
my_set = {10, "Ali", 3.5}
print(my_set)
print(type(my_set))
print(len(my_set))
```

In Python, data structures like lists, tuples, sets, and dictionaries can store multiple data types.
