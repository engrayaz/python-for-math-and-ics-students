```python
import math
```

```python
value1=0.5
value2=0.866
value3=0.577
sininv=math.asin(value1)
cosinv=math.acos(value2)
taninv=math.atan(value3)
print(sininv)
print(cosinv)
print(taninv)
print(math.degrees(sininv))
print(math.degrees(cosinv))
print(math.degrees(taninv))

```
