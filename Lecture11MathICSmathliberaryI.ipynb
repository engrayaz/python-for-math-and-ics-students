```python
# How to import and use liberary in python
"""
A library is a collection of pre-written code that we can reuse.
 Instead of writing everything from scratch, Python provides libraries for:
Math calculations 
Graphics 
Data analysis 
Machine learning 
 Example:
math → math provides functions for real-number mathematical operations like trigonometry, logarithms, and constants.
matplotlib.pyplot → matplotlib.pyplot is used for plotting graphs and visualizing data.
cmath→cmath is used for mathematical operations involving complex numbers.
numpy → advanced math 

Python provides a built-in library called math for mathematical operations. The math library is a built-in Python module that provides:
Trigonometric functions 
Logarithmic functions 
Constants like π (pi) and e
How to Import the library
We use the keyword import to bring a library (module) into our program so we can use its functions. Eg import math"""
```

```python
# import them
import math
a=math.pow(2,3)
print(a)
b=math.pow(16,0.5)
print(b)
c=math.sqrt(16)
print(c)
```

```python
# ln where base e
# log10 base of 10
# log with custom base
d=150
ee=math.log(d)
print(ee)
ef=math.log10(150)
print(ef)
eg=math.log(d,2)
print(eg)
```

```python
eh=math.exp(ee)
ej=10**ef
ek=math.pow(2,eg)
print(eh)
print(ej)
print(ek)
```

```python
# Step 1: Menu
print("Welcome to the Python Calculator!")
print("Select operation:")
print("1. Add (+)")
print("2. Subtract (-)")
print("3. Multiply (*)")
print("4. Exponent (**)")
print("5. Divide (/)")

choice = input("Enter choice (1/2/3/4/5): ")

# Step 2: Input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 3: Conditions (NO nested if)
if choice == '1':
    result = num1 + num2

elif choice == '2':
    result = num1 - num2

elif choice == '3':
    result = num1 * num2

elif choice == '4' and num1 >= 0 and num2 > 0:
    result = num1 ** num2

elif choice == '4':
    result = "Please enter base ≥ 0 and exponent > 0."

elif choice == '5' and num2 != 0:
    result = num1 / num2

elif choice == '5' and num2 == 0:
    result = "Error! Division by zero."

else:
    result = "Invalid Input!"

# Step 4: Output
print("Result:", result)
```

```python
# Step 1: Menu
print("Welcome to the Python Calculator!\n")
print("Select operation:")
print("1. Add (+)")
print("2. Subtract (-)")
print("3. Multiply (*)")
print("4. Exponent (**)")
print("5. Divide (/)")

choice = input("Enter choice (1/2/3/4/5): ")

# Step 2: Input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 3: Conditions
if choice == '1':
    result = num1 + num2

elif choice == '2':
    result = num1 - num2

elif choice == '3':
    result = num1 * num2

elif choice == '4':
    # Restrict to positive numbers only
    if num1 > 0 and num2 > 0:
        result = num1 ** num2
    else:
        result = "Please enter positive numbers only for exponent."

elif choice == '5':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."

else:
    result = "Invalid Input!"

# Step 4: Output
print("Result:", result)
```
