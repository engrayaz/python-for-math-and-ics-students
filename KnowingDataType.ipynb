In the previous lecture, we learned about data types.
Now the question is:
How can we check the data type of a variable in Python?
We use  type() Function
Python provides a built-in function called:
 type()
It tells us the data type of any value or variable.

```python
x=-10
print(type(x))
```

```python
y=-10.5
print(type(y))
```

```python
lang_name="Python"
print(type(lang_name))
```

```python
status=False
print(type(status))
```

```python
z=5-4j
print(type(z))
```

```python
print(type(100))
print(type("Python"))
print(type(False))
```

Recap:  type() is a built-in function ,It tells the data type of a variable ,Very useful for beginners
