Type casting means manually converting one data type into another using built-in functions.

Common Functions:
int()   -> converts to integer
float() -> converts to float
str()   -> converts to string

```python
x1=5
y1=float(x1)
print(type(x1))
print(type(y1))
```

```python
a1="25"
b1=float(a1)
print(type(a1))
print(type(b1))
```

```python
# convert into integer
x2=5.9
y2=int(x2)
print(type(x2))
print(type(y2))
print(y2)
```

```python
a2="25"
b2=int(a2)
print(type(a2))
print(type(b2))
```

```python
# convert to srring
x3=25.7
y3=str(x3)
print(type(x3))
print(type(y3))
```

```python
x4=-12
y4=str(x4)
print(type(x4))
print(type(y4))
```
