
A dictionary stores data in the form of:
Key : Value
Think of a real-life dictionary:
Word → Meaning
Similarly, in Python:
Student Name → Marks
Product Name → Price
Employee ID → Salary
Phone Contact → Number





Create dictionaries in Python,
Access values using keys,
Add new key-value pairs

```python
dict1={"Ali":90,"Sara":80,"John":85}
```

```python
print(dict1)
```

```python
dict2=dict(Ali=90,Sara=80,John=85)
```

```python
print(dict2)
```

```python
dictionary1={101:85,"name":"Ali","Complex":5-3j,1001:"Signals & Systems"}
```

```python
#How to access values, dictionary[key]
sarsMarks=dict2["Sara"]
Complexn=dictionary1["Complex"]
codN=dictionary1[101]
```

```python
print(sarsMarks)
print(Complexn)
print(codN)
```

```python
#Add ne key-value pair, dictionary_name[key] = value
dictionary1["Marks"]=90
dictionary1[10012]=70
```

```python
print(dictionary1)
```
