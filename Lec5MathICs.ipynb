```python
What are Conditional Statements?
Conditional statements are used to run code only when a condition is true.
In real life, we think like this:
If it is raining → take an umbrella
If marks are good → student passes
Else → student fails
In Python, we use:if,if else,if elif else
```

```python
The if Statement Syntax:
if condition:
    # code runs if condition is True
```

```python
x=int(input("enter any number:"))
if x>=0:
    print("the numebr is Positive")
```

```python
We use if else when there are two cases.
Syntax:
if condition:
    # runs if condition is True
else:
    # runs if condition is False
    
```

```python
x=float(input("enter any number:"))
if x>=0:
    print("The number is Positive")
else:
    print ("The number is Negative")
```

```python
#We use elif when there are more than two cases.
Syntax:
if condition1:
    # runs if condition is True
elif condition2:
    # runs if condition2 is True
else:
    # runs if condition1 and 2 are False
```

```python
marks=float(input("enter marks for the student:"))
if marks>=90:
    print("The student achieved A grade")
elif marks>=80:
    print("The student achieved -A grade")
elif marks>=70:
    print("The student achieved B grade")
else:
    print("The student achieved not good grade")
    
```

```python
Important Points
if, elif, and else are used for decision making.
Conditions use comparison operators and logical operators.
Python uses indentation (spaces) to know which code is inside if or else.
 Summary (Lecture 5)
Conditional statements are used to make decisions in programs.
if runs code when a condition is True.
else runs code when the condition is False.
elif is used for multiple conditions.
We used these built-in functions:
print() → to display output
input() → to take input from user
float() → to convert input into decimal 
int() → to convert input into integers 









```

```python
#example on logical operators
# and operator
marks = float(input("Enter your attendance percentage: "))
attendance = float(input("Enter your attendance percentage: "))
if marks >= 50 and attendance >= 75:
    print("You passed")
else:
    print("You failed")
Both conditions must be True to pass.
```

```python
#or operator
num = int(input("Enter a number: "))

if num < 0 or num > 100:
    print("Number is outside the range 0 to 100")
else:
    print("Number is inside the range")
```

```python
# not operator
num = int(input("Enter a number: "))
if not (num == 10):
    print("Number is not 10")
else:
    print("Number is 10")
```
