```python
# My First Python Program: Finding Roots

## What is a Quadratic Equation?

A quadratic equation is an equation of the form:

ax^2 + bx + c = 0

where a â‰  0.

## Quadratic Formula

x1 = (-b + sqrt(b^2 - 4ac)) / (2a)

x2 = (-b - sqrt(b^2 - 4ac)) / (2a)

The two solutions or roots are:

x = x1  or  x = x2

-------------------------------------

Example 1
x^2 - 5x + 6 = 0, has roots/values of x are 3 and2
Example 2
x^2 - 4x + 4 = 0

Factorizing:

(x - 2)^2 = 0

values/Root:

x = 2

(This equation has one repeated root.)


-------------------------------------

Example 2

x^2 - 7x + 10 = 0

Factorizing:

(x - 5)(x - 2) = 0

Roots:

x = 5  and  x = 2


-------------------------------------

Homework

Write python code to solve the following equation using the quadratic formula:

5x^2 - 4x + 4 = 0
```
