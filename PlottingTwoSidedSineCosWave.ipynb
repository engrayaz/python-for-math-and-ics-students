```python
import math
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(-180,180):
  angle_rad=math.radians(i)
  sinv=math.sin(angle_rad)
  y.append(sinv)
  x.append(angle_rad)

plt.plot(x,y)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(-360,360):
  angle_rad=math.radians(i)
  sinv=math.sin(angle_rad)
  y.append(sinv)
  x.append(angle_rad)

plt.plot(x,y)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
x=[]
y=[]
n=int(input("enter number of cycles on both positive and negative sides"))
for i in range(-360*n,360*n):
  angle_rad=math.radians(i)
  sinv=2*math.sin(angle_rad)
  y.append(sinv)
  x.append(angle_rad)

plt.plot(x,y)
plt.show()
```
