```python
# Stpe I
print(" Welcome to the pyton calculaor")
print("Select Operations:")
print(" for addition enter 1")
print(" for subtraction enter 2")
print(" for multiplication enter 3")
print(" for exponent enter 4")
print(" for division enter 5")
# Step 2
choice= input("Enter choice (1/2/3/4/5)")
num1=float(input("enter first number"))
num2=float(input("enter second number"))
# Step III
if choice=='1':
    result=num1+num2
elif choice=='2':
    result=num1-num2
elif choice=='3':
    result=num1*num2
elif choice=='4'and num1>=0 and num2>0:
    result=num1**num2
elif choice=='4':
    result="Please enter num1>=0 and exponent>0"
elif choice=='5'and num2!=0:
    result=num1/num2
elif choice=='5':
    result="Error! Division by zero"
else:
    result="Invalid input"
#Step display result
print(result)
```

```python
# Step 1: Menu
print("Welcome to the Python Calculator!")
print("Select operation:")
print("1. Add (+)")
print("2. Subtract (-)")
print("3. Multiply (*)")
print("4. Exponent (**)")
print("5. Divide (/)")

choice = input("Enter choice (1/2/3/4/5): ")

# Step 2: Input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 3: Conditions (NO nested if)
if choice == '1':
    result = num1 + num2

elif choice == '2':
    result = num1 - num2

elif choice == '3':
    result = num1 * num2

elif choice == '4' and num1 >= 0 and num2 > 0:
    result = num1 ** num2

elif choice == '4':
    result = "Please enter base ≥ 0 and exponent > 0."

elif choice == '5' and num2 != 0:
    result = num1 / num2

elif choice == '5' and num2 == 0:
    result = "Error! Division by zero."

else:
    result = "Invalid Input!"

# Step 4: Output
print("Result:", result)
```

```python
# Step 1: Menu
print("Welcome to the Python Calculator!\n")
print("Select operation:")
print("1. Add (+)")
print("2. Subtract (-)")
print("3. Multiply (*)")
print("4. Exponent (**)")
print("5. Divide (/)")

choice = input("Enter choice (1/2/3/4/5): ")

# Step 2: Input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 3: Conditions
if choice == '1':
    result = num1 + num2

elif choice == '2':
    result = num1 - num2

elif choice == '3':
    result = num1 * num2

elif choice == '4':
    # Restrict to positive numbers only
    if num1 > 0 and num2 > 0:
        result = num1 ** num2
    else:
        result = "Please enter positive numbers only for exponent."

elif choice == '5':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."

else:
    result = "Invalid Input!"

# Step 4: Output
print("Result:", result)
```
