```python
Today we will learn the for loop. To use the for loop for counting,
we will use one important built-in function called range(). 
The range() function gives numbers to the loop, and the loop runs once for each number.
We will use print() to see the output on the screen
Loops are used when we want to repeat some work again and again.

A loop is used to run the same code multiple times.
In real life:
Print numbers from 1 to 10
Ask students’ marks for many students
Print a table of 2
Repeat a message 5 times ie math class, math class, math class, math class, math class
Instead of writing the same code again and again, we use a loop.                   
```

```python
Types of Loops in Python are for loop and while loop. 
In this lecture, we will learn: for loop, in next class we learn while loop
The for loop is used when we know in advance how many times we want to repeat.              
```

```python
Syntax:
for i in range(start, stop,incrment):  
    # code to repeat
Example 4: Print numbers from 1 to 5 using for loop
```

```python
#Example : Print a message 5 times
for i in range(1,6,1):
    print(i)
```

```python
#Example: Print square of numbers from 1 to 5
for b in range(1,6):
    print(b**2)
```

```python
#print table of 5 upto 40
for i in range(5,41,5):
    print(i)
```
